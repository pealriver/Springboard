```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
import pickle
```

```python
# read json into Python
with open(r"C:\A\Education\Data Science\Ultimate Technologies Inc\ultimate_data_challenge.json") as json_file:
    json_data=json.load(json_file)
```

```python
# create the following empty lists
city=[]
trips_in_first_30_days=[]
signup_date=[]
avg_rating_of_driver=[]
avg_surge=[]
last_trip_date=[]
phone=[]
surge_pct=[]
ultimate_black_user=[]
weekday_pct=[]
avg_dist=[]
avg_rating_by_driver=[]
```

```python
# get values frome the dictionries and apend them into respective list
for i,k in enumerate(json_data):
    b=json_data[i]
    row=b.values()
    row=list(row)
    city.append(row[0])
    trips_in_first_30_days.append(row[1])
    signup_date.append(row[2])
    avg_rating_of_driver.append(row[3])
    avg_surge.append(row[4])
    last_trip_date.append(row[5])
    phone.append(row[6])
    surge_pct.append(row[7])
    ultimate_black_user.append(row[8])
    weekday_pct.append(row[9])
    avg_dist.append(row[10])
    avg_rating_by_driver.append(row[11])
```

```python
print(city[:5])
```

```python
# create the dataframe
df=pd.DataFrame()
df['city']=city
df['trips_in_first_30_days']=trips_in_first_30_days
df['signup_date']=signup_date
df['avg_rating_of_driver']=avg_rating_of_driver
df['avg_surge']=avg_surge
df['last_trip_date']=last_trip_date
df['phone']=phone
df['surge_pct']=surge_pct
df['ultimate_black_user']=ultimate_black_user
df['weekday_pct']=weekday_pct
df['avg_dist']=avg_dist
df['avg_rating_by_driver']=avg_rating_by_driver
```

```python
# convert the column of last_trip_date to daytime object
df['last_trip_date'] =  pd.to_datetime(df['last_trip_date'],infer_datetime_format=True)
```

```python
# Create features for last_trip_year, last_trip_month, last_trip_day 
df['last_trip_year'] = df['last_trip_date'].dt.year 
df['last_trip_month'] = df['last_trip_date'].dt.month 
df['last_trip_day'] = df['last_trip_date'].dt.day 
```

```python
df['last_trip_month'].value_counts()
```

```python
# the cut off day of this report is 1st,july,2014, therefore, the active users are those users whose last trips are
# between 1st june and 1st july
df[df['last_trip_month']==7]
```

```python
# create active user feature, if last_trip_month is 6 or 7, then the value will be yes, if last_trip_month is others
# then the value will be no 
active_user=[]
for index, row in df.iterrows():
    if row['last_trip_month']==6 or row['last_trip_month']==7:
        active_user.append('yes')
    else:
        active_user.append('no')
df['active_user']=active_user
```
